# Task 26 Signal Field Analysis

This notebook visualizes continuous ideology signals and community-level vector fields produced by `scripts/analyze_b6_signal_field.py`.

```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt

base = Path("../outputs/task26_signal_analysis_2month")
signals = pd.read_csv(base / "continuous_signals.csv")
monthly = pd.read_csv(base / "community_monthly_means.csv")
velocity = pd.read_csv(base / "community_velocity.csv")
summary = json.loads((base / "vector_field_summary.json").read_text())

print("signals:", signals.shape)
print("monthly:", monthly.shape)
print("velocity:", velocity.shape)
print("months:", summary.get("months", []))
```

```python
latest_month = monthly[monthly["month"] != "unknown"]["month"].max()
snap = monthly[monthly["month"] == latest_month]

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(snap["econ_cont_mean"], snap["social_cont_mean"], s=12, alpha=0.6)
ax.axhline(0.0, color="gray", linewidth=1)
ax.axvline(0.0, color="gray", linewidth=1)
ax.set_title(f"Community Mean Signals ({latest_month})")
ax.set_xlabel("Economic Signal")
ax.set_ylabel("Social Signal")
plt.show()
```

```python
latest_vel_month = velocity["to_month"].max()
vel = velocity[velocity["to_month"] == latest_vel_month]

fig, ax = plt.subplots(figsize=(8, 6))
ax.quiver(
    vel["from_econ"],
    vel["from_social"],
    vel["delta_econ"],
    vel["delta_social"],
    angles="xy",
    scale_units="xy",
    scale=1.0,
    width=0.003,
)
ax.axhline(0.0, color="gray", linewidth=1)
ax.axvline(0.0, color="gray", linewidth=1)
ax.set_title(f"Community Velocity Field ({latest_vel_month})")
ax.set_xlabel("Economic Signal")
ax.set_ylabel("Social Signal")
plt.show()

summary.get("divergence_curl_by_month", {})
```
